```python
def sequential_search(lst, item):

    for pos in range(len(lst)):
        if lst[pos] == item:
            return True 
    
    return False 
```

```python
test_list = [1, 2, 32, 8, 17, 19, 42, 13, 0]

print(sequential_search(test_list, 3))
print(sequential_search(test_list, 13))
```

```python
def ordered_sequential_search(lst, item):
    for pos in range(len(lst)):
        if lst[pos] == item:
            return True 
        if lst[pos] > item:
            return False 

    return False
```

**Binary Search:** Divide-and-Conquer strategy

```python
def binary_search(lst, item):
    first = 0
    last = len(lst) - 1

    while first <= last:
        midpoint = (first + last) // 2
        if lst[midpoint] == item:
            return True 
        elif item < lst[midpoint]:
            # item cannot be in second half of list
            last = midpoint - 1
        else:
            first = midpoint + 1

    return False
```

```python
test_list = [0, 1, 2, 8, 13, 17, 19, 32, 42]

print(binary_search(test_list, 3))
print(binary_search(test_list, 32))
```

**Recursive version**

```python
def binary_search_rec(lst, item):
    # base case
    if len(lst) == 0:
        return False 
    # next line: slightly different... midpoint = (len(lst) - 1) // 2 would match last function
    midpoint = len(lst) // 2 
    if lst[midpoint] == item:
        return True 
    elif item < lst[midpoint]:
        return binary_search_rec(lst[:midpoint], item)
    else:
        return binary_search_rec(lst[midpoint+1:], item)
```

```python
test_list = [0, 1, 2, 8, 13, 17, 19, 32, 42]

print(binary_search_rec(test_list, 3))
print(binary_search_rec(test_list, 32))
```
