```python
def move_tower(height, from_pole, to_pole, with_pole):
    if height < 1:
        return
    move_tower(height - 1, from_pole, with_pole, to_pole)
    move_disk(from_pole, to_pole)
    move_tower(height - 1, with_pole, to_pole, from_pole)

def move_disk(from_pole, to_pole):
    print(f"moving disk from {from_pole} to {to_pole}")
```

```python
move_tower(3, "A", "C", "B")
```

```python
class Disk:
    def __init__(self, label):
        self.label = label 

    def __str__(self):
        return str(self.label)

    def __repr__(self):
        return str(self.label)


class Pole:
    def __init__(self, label):
        self.label = label 
        self.stack = []

    def push(self, disk):
        self.stack.append(disk)
    
    def pop(self):
        return self.stack.pop()


class Tower:
    def __init__(self, height):
        self.height = height 
        self.pole_A = Pole("A")
        self.pole_B = Pole("B")
        self.pole_C = Pole("C")

        for k in range(height, 0, -1):
            self.pole_A.push(Disk(k))

    def move_disk(self, from_pole, to_pole):
        disk = from_pole.pop()
        print(f"moving disk {disk.label} from {from_pole.label} to {to_pole.label}")
        to_pole.push(disk)

    def move_tower(self, height, from_pole, to_pole, with_pole):
        if height < 1:
            return None 
        self.move_tower(height-1, from_pole, with_pole, to_pole)
        self.move_disk(from_pole, to_pole)
        self.move_tower(height-1, with_pole, to_pole, from_pole)

    def solve_puzzle(self):
        self.move_tower(self.height, self.pole_A, self.pole_C, self.pole_B)
        
```

```python
x = Tower(3)
```

```python
print(x.pole_A.stack)
print(x.pole_B.stack)
print(x.pole_C.stack)
```

```python
x.solve_puzzle()
```
