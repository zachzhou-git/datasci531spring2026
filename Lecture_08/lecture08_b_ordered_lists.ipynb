# Ordered (Linked) Lists

<font size = "4">

- An **ordered list** is a collection of items where the relative positions of the items are based on some underlying characteristic.

- For lists of numbers or strings, we can use the `<` comparison operator to determine order.

- The data structure is the same, but we guarantee that the data contained in the nodes remains ordered:

- The methods remain roughly the same, but we should implement them so that we take advantage of the order structure.

```python
from datasci531 import Node
```

```python
class OrderedList:
    def __init__(self):
        # same as unordered
        self._head = None

    def is_empty(self):
        # same as unordered
        return self._head == None
    
    def size(self):
        # same as unordered
        current = self._head
        count = 0 
        while current is not None:
            count = count + 1 
            current = current.next 
        return count

    def search(self, item):
        current = self._head 
        while current is not None:
            if current.data == item:
                return True 
            if current.data > item:
                return False 
            current = current.next 
        return False

    def add(self, item):
        current = self._head 
        previous = None 
        temp = Node(item)

        while current is not None and current.data < item:
            previous = current 
            current = current.next 
        
        if previous is None:
            # should go at beginning of list (smallest element)
            temp.next = self._head 
            self._head = temp 
        else:
            temp.next = current 
            previous.next = temp 

    def remove(self, item):
        current = self._head 
        previous = None 
        found_val = False

        while current is not None and current.data <= item:
            if current.data == item:
                found_val = True
                break
            previous = current 
            current = current.next 

        if not found_val:
            raise ValueError(f"{item} not in list")
            
        if previous is None:
            # Item was found in the first node
            self._head = current.next
        else:
            previous.next = current.next 


    def index(self, item):
        current = self._head 
        ind = -1 
        while current is not None and current.data <= item:
            ind = ind + 1
            if current.data == item:
                return ind 
            current = current.next

        raise ValueError(f"{item} is not in list")

    def pop(self, index = -1):
        if self.is_empty():
            raise IndexError("pop from empty list")

        current = self._head
        previous = None 
        curr_ind = 0

        while current.next is not None:
            if curr_ind == index:
                break
            previous = current
            current = current.next
            curr_ind += 1

        # current.next is None --> at the end of list
        # previous is None --> at beginning of list
        if current.next is None and previous is None:
            # one-element list
            if index == 0 or index == -1:
                # pop only element
                self._head = None
                return None
            else:
                raise IndexError("pop index out of range")

        # reached end of list without break happening
        if current.next is None:
            if curr_ind == index or index == -1:
                previous.next = None 
                return None 
            else:
                raise IndexError("pop index is out of range")


        # OR

        else:
            # it reached the break statement
            if previous is None:
                # pop the first element
                self._head = current.next 
            else:
                previous.next = current.next

    
```

### Computational Complexity of Linked Lists

<font size = "3">

Suppose a linked list has $n$ nodes.

- Cost of `is_empty` = $\mathcal{O}(1)$

- Cost of `size` = $\mathcal{O}(n)$

- Cost of `add` for unordered list = $\mathcal{O}(1)$

- Cost of `append` for unordered list = $\mathcal{O}(n)$

- Worst case cost of `add` for ordered list = $\mathcal{O}(n)$

- Worst case cost of `search` for either list = $\mathcal{O}(n)$

- Worst case cost of `remove` for either list = $\mathcal{O}(n)$

**Note:** Python "lists" must be some other data structure, because for a linked list we have:

- Cost of `pop()` = $\mathcal{O}(n)$

- Cost of `pop(0)` = $\mathcal{O}(1)$

which is the opposite for Python lists!
